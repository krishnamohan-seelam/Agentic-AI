## Implement the Streaming Function

Complete the function below to stream a response using LangChain.

```python
def stream_response(prompt: str) -> str:
    """
    Stream an LLM response using LangChain.

    Args:
        prompt: The user's prompt

    Returns:
        The complete response as a string
    """
    print("Assistant: ", end="", flush=True)

    # TODO 1: Initialize a ChatOpenAI model with model="gpt-4o-mini" and streaming=True

    # TODO 2: Use llm.stream(prompt) to get a streaming response
    # Iterate through the chunks, print chunk.content in real-time, and accumulate the full response

    full_response = ""

    # Don't have access to gpt-4-1106-preview using gpt-4o-mini
    llmodel= ChatOpenAI(model="gpt-4o-mini",api_key=OPEN_AI_API_KEY, streaming=True)
    for chunk in llmodel.stream(prompt):
        print(chunk.content, end="", flush=True)
        full_response += chunk.content

    print()  # New line at the end
    return full_response
```

## Run Your Implementation

Test your streaming function with different prompts.
