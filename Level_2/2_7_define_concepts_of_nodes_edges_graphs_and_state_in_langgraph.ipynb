## Setup

Run this cell to import all required libraries.

```python
# Setup - run this cell first

from typing import TypedDict
from langgraph.graph import StateGraph, START, END

print("Setup complete!")
```

## Context

You are building a text processing pipeline that takes a message and transforms it through two stages:

1. **Uppercase Node**: Converts the text to all uppercase letters
2. **Prefix Node**: Adds a prefix like "[PROCESSED] " to the beginning of the text

**Input:** A state containing:
- `text`: The original message (e.g., "hello world")
- `prefix`: The prefix to add (e.g., "[ALERT] ")

**Expected Output:** After running the graph with `text="hello world"` and `prefix="[ALERT] "`:
- `text` should be `"[ALERT] HELLO WORLD"`
- `prefix` should remain `"[ALERT] "`

## Step 1: Define the State

Create a State class using TypedDict that holds the data for your pipeline.

```python
class TextState(TypedDict):
    """
    State for the text processing pipeline.

    Fields:
        text: The text being processed (will be transformed by nodes)
        prefix: The prefix to add to the text (read by prefix node)
    """
    # TODO: Define the two fields with their types
    text:str = "hello world"
    prefix:str = "[ALERT] "
```

## Step 2: Create Node Functions

Create two node functions that read from state and return updates.

```python
def to_uppercase(state: TextState) -> dict:
    """
    Convert the text field to uppercase.

    Args:
        state: Current state containing 'text' field

    Returns:
        Dictionary with updated 'text' field (uppercase version)
    """
    # TODO: Read text from state, convert to uppercase, return as dict
    state["text"] = state["text"].upper()
    return state


def add_prefix(state: TextState) -> dict:
    """
    Add the prefix to the beginning of the text.

    Args:
        state: Current state containing 'text' and 'prefix' fields

    Returns:
        Dictionary with updated 'text' field (prefix + text)
    """
    # TODO: Read text and prefix from state, combine them, return as dict
    state["text"] =  state["prefix"] +  state["text"]
    return state
```

## Step 3: Build and Compile the Graph

Create the graph builder, add nodes and edges, then compile it.

```python
def build_text_pipeline() -> object:
    """
    Build and compile the text processing graph.

    The graph should execute in this order:
    START -> to_uppercase -> add_prefix -> END

    Returns:
        Compiled graph ready for invocation
    """
    # TODO 1: Create a StateGraph with TextState
    builder = StateGraph(TextState)
    # TODO 2: Add the two nodes to the graph
    builder.add_node("to_uppercase",to_uppercase)
    builder.add_node("add_prefix",add_prefix)
    # TODO 3: Add edges to define the flow (START -> uppercase -> prefix -> END)
    builder.add_edge(START,"to_uppercase")
    builder.add_edge("to_uppercase","add_prefix")
    builder.add_edge("add_prefix",END)
    # TODO 4: Compile and return the graph
    graph = builder.compile()
    return graph
```

## Run Your Implementation

Test your graph with sample inputs.

```python
# Build the graph

graph = build_text_pipeline()

# Test 1: Basic test
result = graph.invoke({
    "text": "hello world",
    "prefix": "[ALERT] "
})
print(f"Test 1 - Input: 'hello world' with prefix '[ALERT] '")
print(f"Output: {result['text']}")
print(f"Expected: [ALERT] HELLO WORLD")
print()
```
