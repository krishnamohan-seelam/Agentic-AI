## Your Task

Implement an async function that translates text into multiple languages **concurrently** using `llm.ainvoke()` and `asyncio.gather()`.

**Input:**
- `text`: A string to translate
- `languages`: A list of target languages

**Output:** A list of dicts with `language` and `translation` keys

```python
async def translate_concurrently(text: str, languages: list[str]) -> list[dict]:
    """
    Translate text into multiple languages concurrently.

    Args:
        text: The text to translate
        languages: List of target languages

    Returns:
        List of dicts with 'language' and 'translation' keys
    """
    # TODO: Implement concurrent translation
    # 1. Create a list of coroutines using llm.ainvoke() for each language
    #    Hint: prompt = f"Translate '{text}' to {language}. Return only the translation."
    # 2. Execute all concurrently with await asyncio.gather()
    # 3. Combine languages with results into list of dicts
    tasks = [ llm.ainvoke(input=f"Translate '{text}' to {language}") for language in languages]
    async_results = await asyncio.gather(*tasks)
    return [{"language": lang, "translation": result.content} for lang, result in zip(languages,async_results)]
```

## Test Your Implementation

```python
text = "Hello, how are you today?"
languages = ["Spanish", "French", "German", "Japanese"]

results = await translate_concurrently(text, languages)

for r in results:
    print(f"{r['language']}: {r['translation']}")
```
