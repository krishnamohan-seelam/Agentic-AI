## Part 1: Product Summary Template

You are building a tool that generates product summaries for an e-commerce platform. The summary needs to be customizable based on:
- **product_name**: The name of the product (e.g., "Wireless Headphones")
- **category**: The product category (e.g., "Electronics", "Home & Garden")
- **tone**: The writing style (e.g., "professional", "casual", "enthusiastic")

Your template should produce a prompt like:
```
Write a professional summary for the product "Wireless Headphones" in the Electronics category. Keep it concise and highlight key selling points.
```

Create a `PromptTemplate` that accepts these three variables and generates a product summary request.

````python
def create_product_summary_template() -> PromptTemplate:
    """
    Create a PromptTemplate for generating product summaries.

    The template must include three variables:
    - {tone}: The writing style for the summary
    - {product_name}: The name of the product
    - {category}: The product category

    Returns:
        PromptTemplate: A template that generates product summary requests
    """
    # TODO: Create and return a PromptTemplate using from_template()
    # The template string should incorporate all three variables
    prompt_template = PromptTemplate.from_template("Write a {tone} summary for the product {product_name} in the {category} category.")
    return prompt_template
````

## Test Your Product Summary Template

Run this cell to test your implementation with sample data.

````python
# Test your template
summary_template = create_product_summary_template()

# Test with sample values
prompt = summary_template.invoke({
    "product_name": "Smart Water Bottle",
    "category": "Health & Fitness",
    "tone": "enthusiastic"
})

print("Formatted Prompt:")
print(prompt.to_string())
print("\n" + "="*60 + "\n")

# Get response from the model
response = llm.invoke(prompt.to_string())
print("Model Response:")
print(response.content)
````
